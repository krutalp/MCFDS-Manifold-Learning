```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.manifold import Isomap
from sklearn.cluster import KMeans
from sklearn import metrics

# Load the dataset
data = pd.read_csv("Datasets/CC GENERAL.csv")

# Drop unnecessary columns
data.drop(["CUST_ID"], axis = 1, inplace = True)

# Replace missing values with column means
data.fillna(data.mean(), inplace = True)

# Create Isomap object
isomap = Isomap(n_components = 2)

# Fit and transform the data with Isomap
isomap_data = isomap.fit_transform(data)
```

```python
# Compute the optimal k value using the elbow method
inertias = []
silhouettes = []
ks = range(2, 11)
for k in ks:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(isomap_data)
    inertias.append(kmeans.inertia_)
    silhouettes.append(metrics.silhouette_score(isomap_data, kmeans.labels_))
    
# Plot the elbow curve to find the optimal k value
plt.plot(ks, inertias, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()

# Plot the silhouette scores to help evaluate the optimal k value
plt.plot(ks, silhouettes, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Optimal k')
plt.show()
```

```python
# Perform k-means clustering on the transformed data with optimal k
optimal_k = 6  # Set to the k value with the elbow method or another method of your choice
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans_clusters = kmeans.fit_predict(isomap_data)

# Create a scatter plot of the transformed data, colored by cluster
plt.scatter(isomap_data[:, 0], isomap_data[:, 1], c=kmeans_clusters)
plt.title(f"K-Means Clustering of Isomap Transformed Credit Card Dataset with {optimal_k} clusters")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.show()
```
